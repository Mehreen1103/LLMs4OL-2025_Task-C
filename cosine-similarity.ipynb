```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Load test types
with open("/kaggle/input/data-ontology/obi_test_types.txt", "r") as f:
    types = [line.strip() for line in f.readlines()]

# Load sentence embedding model (Bio-compatible)
model = SentenceTransformer('all-MiniLM-L6-v2')  # You can change to BioBERT-based ones if needed

# Compute embeddings for all types
embeddings = model.encode(types)

# Threshold for determining is-a relation (you can tune this)
# SIMILARITY_THRESHOLD = 0.82
# SIMILARITY_THRESHOLD = 0.90
SIMILARITY_THRESHOLD = 0.95


# Predict is-a relations
results = []
for i, parent in enumerate(types):
    for j, child in enumerate(types):
        if i == j:
            continue
        sim = cosine_similarity([embeddings[i]], [embeddings[j]])[0][0]
        if sim > SIMILARITY_THRESHOLD:
            results.append({
                "parent": parent,
                "child": child
            })

# Save output JSON
with open("submission.json", "w") as f:
    json.dump(results, f, indent=2)

print(f"✅ Done! Found {len(results)} is-a relations.")
```
